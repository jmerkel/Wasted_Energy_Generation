```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
energy_df=pd.read_csv("Resources/energy_dataset.csv",index_col=0,parse_dates=[0])
energy_df.head()
```

```python
energy_df.index=pd.to_datetime(energy_df.index,utc=True)
energy_df=energy_df['2015':'2018']
energy_df
```

```python
energy_df.columns
```

```python
def hour_columns(data,target_col):
    data.loc[:,'year'] = data.index.year
    data.loc[:,'month'] = data.index.month
    data.loc[:,'day'] = data.index.day
    data.loc[:,'hours'] = data.index.hour
    data.loc[:,'date'] = pd.to_datetime(data.loc[:,['year', 'month', 'day']], format='%Y-%m-%d', errors='ignore')
    
    #set the index to dates only
    data = data.set_index(pd.DatetimeIndex(data['date']))
    #drop non target columns 
    data = data.loc[:,[target_col, 'hours']]
    #pivot the table into the format Date h0, h1, ...h23
    data = data.pivot(columns='hours', values=target_col)
    return data

biomass=hour_columns(energy_df,'generation biomass')
biomass
```

```python
pd.set_option('display.max_columns',None)
```

```python
# The biomass dataframe is composed of the datetime index and other 24 columns from 0-23 hours for each day. 
#The value represent the total energy generation. Combine them into one single column to calculate the 
# total amount of biomass for each day

def total(data):
    data['Total']=data.sum(axis=1)
    return pd.DataFrame(data['Total'])
new_biomass=total(biomass)
new_biomass
```

```python
#Convert the index from datetime into month
new_biomass.index=new_biomass.index.map(lambda x: x.strftime('%Y-%m'))
new_biomass.index.name='hours'
```

```python
new_biomass=new_biomass.sum(level='hours')
```

```python
biomass_2015=pd.DataFrame(new_biomass['Total'][0:12])
biomass_2016=pd.DataFrame(new_biomass['Total'][12:24])
biomass_2017=pd.DataFrame(new_biomass['Total'][24:36])
biomass_2018=pd.DataFrame(new_biomass['Total'][36:48])
biomass_2015
```

```python
month=['January','February','March','April','May','June','July','August','September','October','November','December']
month
```

```python
biomass_2015['Total']
```

```python
plt.plot(month,biomass_2015['Total'],label="2015",marker='.')
plt.plot(month,biomass_2016['Total'],label="2016",marker='.')
plt.plot(month,biomass_2017['Total'],label="2017",marker='.')
plt.plot(month,biomass_2018['Total'],label="2018",marker='.')
plt.xticks(rotation='vertical',fontsize=12)
plt.xlabel("Month",fontsize=18)
plt.ylabel("Biomass Genearation",fontsize=18)
plt.title("Biomass Generation(2015-2018)",fontsize=20)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0., prop={'size': 10})
fig=plt.gcf()
fig.set_size_inches(18.5,10.5)
fig.savefig('biomassgeneration.png',dpi=100)
plt.show()
```
